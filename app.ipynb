```python
import cv2  
import easygui 
import numpy as np 
import imageio
```

```python
import sys
import matplotlib.pyplot as plt
import os
import tkinter as tk
from tkinter import filedialog
from tkinter import *
from PIL import ImageTk, Image
```

```python
top = tk.Tk()
top.geometry('400x500')
top.title('Cartoonify Your Image !')
top.resizable(False, False)
img = Image.open('evenement_culturel.png')
bg = ImageTk.PhotoImage(img)
label = Label(top, image=bg, bd=0)
label.place(x=0, y=0)
```

```python
def upload():
    ImagePath = easygui.fileopenbox()
    cartoonify(ImagePath)
```

```python
def cartoonify(ImagePath):
    # read the image
    originalmage = cv2.imread(ImagePath)
    originalmage = cv2.cvtColor(originalmage, cv2.COLOR_BGR2RGB)
  

    
    if originalmage is None:
        print("Can not find any image. Choose appropriate file")
        sys.exit()

    ReSized1 = cv2.resize(originalmage, (960, 540))
   

    
    grayScaleImage = cv2.cvtColor(originalmage, cv2.COLOR_BGR2GRAY)
    ReSized2 = cv2.resize(grayScaleImage, (960, 540))
   

    
    smoothGrayScale = cv2.medianBlur(grayScaleImage, 5)
    ReSized3 = cv2.resize(smoothGrayScale, (960, 540))
    

   
    getEdge = cv2.adaptiveThreshold(smoothGrayScale, 255,
                                    cv2.ADAPTIVE_THRESH_MEAN_C,
                                    cv2.THRESH_BINARY, 9, 9)

    ReSized4 = cv2.resize(getEdge, (960, 540))
   

    
    colorImage = cv2.bilateralFilter(originalmage, 9, 300, 300)
    ReSized5 = cv2.resize(colorImage, (960, 540))


   
    cartoonImage = cv2.bitwise_and(colorImage, colorImage, mask=getEdge)

    ReSized6 = cv2.resize(cartoonImage, (960, 540))
  
    images = [ReSized1, ReSized2, ReSized3, ReSized4, ReSized5, ReSized6]

    fig, axes = plt.subplots(3, 2, figsize=(8, 8), subplot_kw={'xticks': [], 'yticks': []},
                             gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(images[i], cmap='gray')

    save1 = Button(top, text="Click Here to Save Cartoon Image", command=lambda: save(ReSized6, ImagePath), padx=10, pady=5)
    save1.configure(background='purple', foreground='white', font=btn_font, width=25, height=1)
    save1.pack(side=TOP, pady=10)

    plt.show()


def save(ReSized6, ImagePath):
    
    newName = "cartoonified_Image"
    path1 = os.path.dirname(ImagePath)
    extension = os.path.splitext(ImagePath)[1]
    path = os.path.join(path1, newName + extension)
    cv2.imwrite(path, cv2.cvtColor(ReSized6, cv2.COLOR_RGB2BGR))
    I = "Image saved by name " + newName + " at " + path
    tk.messagebox.showinfo(title=None, message=I)

from tkinter import Tk, font
btn_font = font.nametofont("TkDefaultFont")

btn_font.configure(family="Segoe Script",
                           size=12,
                           weight=font.BOLD)

upload = Button(top, text="Click Here to Upload an Image", command=upload, padx=10, pady=5)
upload.configure(background='purple', foreground='white', font=btn_font, width=25, height=1)
upload.pack(side=TOP, pady=140)

top.mainloop()
```
